```python
#SCRAPING
#-------------------------------------------------------------------------------------------------------------------------------
import selenium
from selenium import webdriver
import pandas as pd
from pandas import ExcelWriter
from pandas import ExcelFile
from selenium.webdriver.support.ui import Select
import time
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
import os
from datetime import datetime
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

driverPath = r"C:\Users\Merit\PycharmProjects\chrome\chromedriver_win32 (1)\chromedriver.exe"
driver = webdriver.Chrome(executable_path = driverPath)
driver.get("https://www.cravath.com/people/index.html")
print(driver.title)
print(driver.current_url)
driver.maximize_window()
#cache
driver.find_element_by_xpath("(//*[contains(@class, 'icon-Close styles__closeButton--03a22cd7')])").click()
time.sleep(3)
#mainCategoryClass
a = driver.find_elements_by_class_name("l-vgap-generic-interior-section")

first=[]
second=[]
third=[]
forth=[]
fifth=[]
sixth=[]
for j in range(0,len(a)):
    try:
#withview
        if(a[j].find_element_by_class_name("card-lister-cta__viewAllArrow--875950ce")):

            a[j].find_element_by_class_name("card-lister-cta__viewAllArrow--875950ce").click()
            time.sleep(5)
            while (True):
                part = a[j].find_elements_by_class_name("l-content-container")
                card = a[j].find_elements_by_class_name("card-lister__box--554421f0")
                for i in range(0, len(card)):
                   
                    detail = card[i].text.split("\n")

                    name = detail[0]
                    practice = detail[1]
                    index1 = [i for i, elem in enumerate(detail) if '@' in elem]
                    email = detail[index1[0]] if len(index1) else "Null"
                    index2 = [i for i, elem in enumerate(detail) if '+' in elem]
                    tel = detail[index2[0]] if len(index2) else "Null"
                    position = part[0].text.capitalize()

                    first.append(name)
                    second.append(email)
                    third.append(tel)
                    forth.append(practice)
                    fifth.append(position)
                    sixth.append(driver.current_url)
                  
                try:
                    if (driver.find_element_by_xpath(
                            "(//*[contains(@class, 'styles__arrow--fd23c3f5 icon-caret-thin-right hover-color-dodger-blue')])").is_enabled()):
                        driver.find_element_by_xpath(
                            "(//*[contains(@class, 'styles__arrow--fd23c3f5 icon-caret-thin-right hover-color-dodger-blue')])").click()

                        time.sleep(2)
                      
                except:

                    driver.find_element_by_xpath("//li//following::a[contains(text(),'People')][1]").click()
                    break
            driver.refresh()

#withoutview
    except:
        part = a[j].find_elements_by_class_name("l-content-container")
        card = a[j].find_elements_by_class_name("card-lister__box--554421f0")
        for i in range(0, len(card)):
           
            detail = card[i].text.split("\n")
            name = detail[0]
            practice = detail[1]
            index1 = [i for i, elem in enumerate(detail) if '@' in elem]
            email = detail[index1[0]] if len(index1) else "Null"
            index2 = [i for i, elem in enumerate(detail) if '+' in elem]
            tel = detail[index2[0]] if len(index2) else "Null"
            position = part[0].text.capitalize()

            first.append(name)
            second.append(email)
            third.append(tel)
            forth.append(practice)
            fifth.append(position)
            sixth.append(driver.current_url)

    a = driver.find_elements_by_class_name("l-vgap-generic-interior-section")
```

```python
# dataframe
df = pd.DataFrame()
df['Sno'] = df.index + 1
df.set_index('Sno',inplace=True)
df['NAME'] = first
df['EMAIL'] = second
df['TEL'] = third
df['PRACTICE'] = forth
df['POSITION'] = fifth
df['URL'] = sixth
df['DATETIME']= pd.Timestamp(datetime.now())
```

```python
# FlaskAPI with Sqlite

import os
import sqlite3
from flask import jsonify
from flask_api import FlaskAPI
from flask_restful import Api
from sqlalchemy import create_engine
from flask import request
import json

#default route
app=FlaskAPI(__name__)
api=Api(app)
c=create_engine('sqlite:///peopledb.db')

#create   
@app.route('/people_create', methods=['GET'])
def create():
    try:
        con = c.connect()
        trans = con.begin()
        con.execute('CREATE TABLE IF NOT EXISTS peopletbl(SNO INT PRIMARY KEY,NAME TEXT,EMAIL TEXT,TEL TEXT,PRACTICE TEXT,POSTION TEXT,TIMESTAMP DATETIME DEFAULT CURRENT_TIMESTAMP)')
        trans.commit()
        text = {"Message " : "Table Created Successfully"}
        return text
    except Exception as error:
        text = {"Message " : "Table Creation Failed"}
        return text
    finally:
        con.close()

# insert    
@app.route('/people_insert', methods=['GET'])
def insert_tbl():
    con = c.connect()
    df.to_sql('peopletbl', con,if_exists='replace')
    text = {"Message " : "Inserted Successfully"}
    return text

# select
@app.route('/people_select', methods=['GET'])
def select():
    con = c.connect()
    selectdata = con.execute("select * from peopletbl")  
    return json.dumps([dict(x) for x in selectdata])

app.run()
```
